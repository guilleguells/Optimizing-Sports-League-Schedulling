```python
from itertools import product
n_teams = 10
n_minus_1 = n_teams - 1
opt_1 = 5
opt_2 = 4

def valid_pattern(pattern):
    # Count occurrences of 'H' and 'A'
    h_count = pattern.count('H')
    a_count = pattern.count('A')

    # Check for similar number of H and A matches (4H-5A or 5H-4A)
    if not (h_count == opt_1 and a_count == opt_2 or h_count == opt_2 and a_count == opt_1):
        return False

    # Check for more than two consecutive H or A matches
    for i in range(len(pattern) - 2):
        if pattern[i] == pattern[i + 1] == pattern[i + 2]:
            return False

    return True

# Generate all combinations of H and A for N-1 matches
all_patterns = product('HA', repeat=n_minus_1)

# Filter patterns that meet the criteria
valid_patterns = [''.join(pattern) for pattern in all_patterns if valid_pattern(pattern)]

# Output the number of valid patterns
print(f"Total valid patterns: {len(valid_patterns)}")
```

```python
# Function to check if two patterns are complementary
def are_complementary(pattern1, pattern2):
    return all((c1 == 'H' and c2 == 'A') or (c1 == 'A' and c2 == 'H') for c1, c2 in zip(pattern1, pattern2))

# Find all pairs of complementary patterns
C = []
for i in range(len(valid_patterns)):
    for j in range(i + 1, len(valid_patterns)):
        if are_complementary(valid_patterns[i], valid_patterns[j]):
            C.append((valid_patterns[i], valid_patterns[j]))

print(C)

C_indices = [(valid_patterns.index(x), valid_patterns.index(y)) for x, y in C]

print(C_indices)
```

Team definition


```python
# Create the list of teams and ranking
teams = ['Team 1', 'Team 2', 'Team 3', 'Team 4', 'Team 5', 'Team 6', 'Team 7', 'Team 8', 'Team 9', 'Team 10']


ranking = {
    1: teams[:3],  # Category 1
    2: teams[3:6],  # Category 2
    3: teams[6:]    # Category 3
}

# Assign a unique number to each team
team_numbers = {i : team for i, team in enumerate(teams)}

# Create a reverse mapping to look up the number of a team
team_to_number = {team: number for number, team in team_numbers.items()}

# Create a dictionary to store subsets of stronger teams
stronger_teams = {}

# Iterate through each team and determine stronger teams
for position, teams_at_position in ranking.items():
    for team in teams_at_position:
        team_number = team_to_number[team]  # Get the team's unique number
        stronger_teams[team_number] = [
            team_to_number[stronger_team]
            for higher_position in range(1, position)  # Look at higher positions only
            for stronger_team in ranking[higher_position]
        ]

# Set of teams located in the same town
D = [(0,5)]

# Print the team-to-number mapping and the stronger teams dictionary
print("Team Numbers:", team_numbers)
print("Stronger Teams:", stronger_teams)
print('Pairs of teams located in the same town', D)
```

Defining and Solving the Model

```python
from gurobipy import Model, GRB, quicksum

# Initialize the model
m = Model("Team_to_Pattern_Allocation")


T = range(n_teams)       # Teams: i, j ∈ T
P = selected_patterns    # Patterns: p ∈ P
W = range(n_minus_1)     # Weeks: w ∈ W 


# Given calendar: dictionary in the form {(i, j, t): 1.0}
calendar = schedule

# z_{i,p} = 1 if team i is assigned to pattern p 
z_var = m.addVars(T, P, vtype=GRB.BINARY, name="z")

# s_{i,w} = 1 if team i faces strictly stronger opponents in both rounds w and w+1
# We'll consider w from 0..N−2 for consecutive weeks
s_var = m.addVars(T, range(n_minus_1 - 1), vtype=GRB.BINARY, name="s")

# x_{i,j,w} = 1 if team i faces team j in week w 
x_var = m.addVars(T, T, W, vtype=GRB.BINARY, name="x")


# Objective: Minimize carry-over effect
m.setObjective(
    quicksum(s_var[i, w] for i in T for w in range(n_minus_1 - 1)),
    GRB.MINIMIZE
)



# Constraint (4.4b): Each team is assigned to exactly one pattern
for i in T:
    m.addConstr(
        quicksum(z_var[i, p] for p in P) == 1,
        name=f"team_pattern_allocation_{i}"
    )


# Constraint (4.4c): Each selected pattern is assigned to exactly one team
for p in P:
    m.addConstr(
        quicksum(z_var[i, p] for i in T) == 1,
        name=f"unique_pattern_assignment_{p}"
    )


# Same-venue constraints (4.4d): 
# If teams i and j share a venue, and patterns p, p' are complementary,
# then z_{i,p} + z_{i,p'} - (z_{j,p} + z_{j,p'}) = 0

for (i, j) in D:  # teams i, j share a venue
    for (p1, p2) in C_indices:
        if p1 in selected_patterns:
            m.addConstr(
                (z_var[i, p1] + z_var[i, p2]) - (z_var[j, p1] + z_var[j, p2]) == 0,
                name=f"same_venue_{i}_{j}_{p1}_{p2}"
            )


# Constraint (4.4e): Linking x_{i,j,w} to pattern assignments 
# Here 'calendar' is 'Q', a set/dict of pairs (p,p') that meet in week w
# x_{i,j,w} ≥ z_{i,p} + z_{i,p'} + z_{j,p} + z_{j,p'} - 1

for (p1, p2, w) in schedule.keys():  # schedule dict might say: (p1, p2, w) => 1
    for i in T:
        for j in T:
            if i < j:
                m.addConstr(
                    x_var[i, j, w] >= z_var[i, p1] + z_var[i, p2]
                                     + z_var[j, p1] + z_var[j, p2] - 1,
                    name=f"match_link_{i}_{j}_{w}_{p1}_{p2}"
                )


# Constraint (4.4f): Each pair (i,j) meets exactly once
for i in T:
    for j in T:
        if i < j:
            m.addConstr(
                quicksum(x_var[i, j, w] for w in W) == 1,
                name=f"one_match_{i}_{j}"
            )


# Constraint (4.4g): Each team i plays exactly one match per week
# sum_{j in T} x_{i,j,w} = 1
# (with i < j or j < i logic to avoid duplication)

for w in W:
    for i in T:
        m.addConstr(
            quicksum(x_var[i, j, w] for j in T if i < j)
            + quicksum(x_var[j, i, w] for j in T if j < i) == 1,
            name=f"one_match_per_week_{i}_{w}"
        )


# Avoid double counting x_{i,j,w} if i >= j
for i in T:
    for j in T:
        if i >= j:
            for w in W:
                x_var[i, j, w].ub = 0


# Constraints (4.4h)–(4.4j): Carry-over effect
# For each team i and each w in 0..(N-2)-1 => consecutive weeks w, w+1
# s_{i,w} <= sum_{j in S_i} (x_{i,j,w} + x_{j,i,w}), etc.

for i in T:
    for w in range(n_minus_1 - 1):  
        # 4.4h: sum_{j in S_i}(x_{i,j,w} + x_{j,i,w} + x_{i,j,w+1} + x_{j,i,w+1}) <= 1 + s_{i,w}
        m.addConstr(
            quicksum(x_var[i, j, w]   for j in stronger_teams[i] if i < j)
            + quicksum(x_var[j, i, w] for j in stronger_teams[i] if j < i)
            + quicksum(x_var[i, j, w+1]   for j in stronger_teams[i] if i < j)
            + quicksum(x_var[j, i, w+1] for j in stronger_teams[i] if j < i)
            <= 1 + s_var[i, w],
            name=f"coe_upper_{i}_{w}"
        )

        # 4.4i: s_{i,w} <= sum_{j in S_i}(x_{i,j,w} + x_{j,i,w})
        m.addConstr(
            s_var[i, w] <=
            quicksum(x_var[i, j, w] for j in stronger_teams[i] if i < j)
            + quicksum(x_var[j, i, w] for j in stronger_teams[i] if j < i),
            name=f"coe_lower1_{i}_{w}"
        )

        # 4.4j: s_{i,w} <= sum_{j in S_i}(x_{i,j,w+1} + x_{j,i,w+1})
        m.addConstr(
            s_var[i, w] <=
            quicksum(x_var[i, j, w+1] for j in stronger_teams[i] if i < j)
            + quicksum(x_var[j, i, w+1] for j in stronger_teams[i] if j < i),
            name=f"coe_lower2_{i}_{w}"
        )


# Solve the model
m.optimize()


# Retrieve solution if optimal
if m.status == GRB.OPTIMAL:
    print("Optimal solution found.")
    z_sol = {
        (i, p): z_var[i, p].x
        for i in T for p in P
        if z_var[i, p].x > 0.5
    }
    s_sol = {
        (i, w): s_var[i, w].x
        for i in T for w in range(n_minus_1 - 1)
        if s_var[i, w].x > 0.5
    }
    x_sol = {
        (i, j, w): x_var[i, j, w].x
        for i in T for j in T for w in W
        if x_var[i, j, w].x > 0.5
    }
    print("Team-to-pattern allocation (z):", z_sol)
    print("Carry-over effect (s):", s_sol)
    print("Matches between teams (x):", x_sol)
else:
    print("No optimal solution found.")
```

Checking Patterns for 'H' and 'A' Specifications

```python
pattern_to_team = {team_numbers[team] : p for (team, p), value in z_sol.items() if value > 0.5}
weeks = {}

for (team1, team2, week), value in x_sol.items():
    if value == 1.0: 
        if week not in weeks:
            weeks[week] = []
        key = pattern_to_team[team_numbers[team1]]  
        if valid_patterns[key][week] == 'H':
            match = f"{team_numbers[team1]} vs {team_numbers[team2]}"
        else:
            match = f"{team_numbers[team2]} vs {team_numbers[team1]}"
        if match not in weeks[week]:
            weeks[week].append(match)


```
